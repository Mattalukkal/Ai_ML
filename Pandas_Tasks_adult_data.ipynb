Create a new column 'income_binary' where <=50K is 0 and >50K is 1.

```python
df_churn['income_binary'] = df_churn['income'].apply(lambda x: 0 if x == '<=50K' else 1)
display(df_churn.head())
```

Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).

```python
conditions = [
    df_churn['age'] < 30,
    (df_churn['age'] >= 30) & (df_churn['age'] <= 60),
    df_churn['age'] > 60
]
choices = ['Youth', 'Adult', 'Senior']
df_churn['age_group'] = np.select(conditions, choices, default='Unknown')
display(df_churn.head())
```

Rename the 'educational-num' column to 'education_num'.

```python
df_churn.rename(columns={'educational-num': 'education_num'}, inplace=True)
display(df_churn.columns)
```

Replace the missing values represented by '?' in 'workclass' and 'occupation' columns with 'Unknown'.

```python
df_churn['workclass'] = df_churn['workclass'].replace('?', 'Unknown')
df_churn['occupation'] = df_churn['occupation'].replace('?', 'Unknown')
display(df_churn.head())
```

Calculate the average age for each workclass.

```python
average_age_by_workclass = df_churn.groupby('workclass')['age'].mean()
display(average_age_by_workclass)
```

Find the total capital-gain for each gender.

```python
total_capital_gain_by_gender = df_churn.groupby('gender')['capital-gain'].sum()
display(total_capital_gain_by_gender)
```

Get the count of each marital-status category.

```python
marital_status_counts = df_churn['marital-status'].value_counts()
display(marital_status_counts)
```

Find the mean hours-per-week worked by individuals in different education levels.

```python
mean_hours_per_week_by_education = df_churn.groupby('education')['hours-per-week'].mean()
display(mean_hours_per_week_by_education)
```

Calculate the mean, median, and standard deviation of 'fnlwgt' for each race.

```python
fnlwgt_stats_by_race = df_churn.groupby('race')['fnlwgt'].agg(['mean', 'median', 'std'])
display(fnlwgt_stats_by_race)
```

Get the unique values and their counts for the 'native-country' column

```python
native_country_counts = df_churn['native-country'].value_counts()
display(native_country_counts)
```

Filter the dataset to include only individuals with income >50K.

```python
df_income_greater_50k = df_churn[df_churn['income'] == '>50K']
display(df_income_greater_50k.head())
```

Sort the dataset by age in descending order.

```python
df_churn_sorted_by_age = df_churn.sort_values(by='age', ascending=False)
display(df_churn_sorted_by_age.head())
```

Filter the dataset to include only individuals who work more than 40 hours per week.



```python
df_hours_per_week_greater_40 = df_churn[df_churn['hours-per-week'] > 40]
display(df_hours_per_week_greater_40.head())
```

Calculate the percentage of missing values in each column.

```python
missing_values_percentage = (df_churn.isnull().sum() / len(df_churn)) * 100
display(missing_values_percentage)
```

Calculate the median income by education level and gender.

```python
median_income_by_education_gender = df_churn.groupby(['education', 'gender'])['income_binary'].median()
display(median_income_by_education_gender)
```

Find the top 3 occupations with the highest average capital-gain.

```python
top_3_occupations_capital_gain = df_churn.groupby('occupation')['capital-gain'].mean().nlargest(3)
display(top_3_occupations_capital_gain)
```

# Chart 1 : Distribution of Age Groups

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df_churn, x='age_group')
plt.title('Distribution of Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

# Observation : The 'Adult' age group (30-60) is significantly larger than 'Youth' (< 30) and 'Senior' (60+), indicating that the dataset is predominantly composed of working-age individuals.

# Chart 2 : Income Distribution by Gender

```python
plt.figure(figsize=(10, 7))
sns.countplot(data=df_churn, x='income_binary', hue='gender', palette='viridis')
plt.title('Income Distribution by Gender')
plt.xlabel('Income (0: <=50K, 1: >50K)')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```

# Observation: The plot clearly shows that for both income categories (<=50K and >50K), there are significantly more males than females.
# Specifically, a larger proportion of males are in the >50K income bracket compared to females, suggesting a potential gender disparity in higher income earnings.


# Chart 3 : Distribution of Workclass Categories

```python
plt.figure(figsize=(12, 7))
sns.countplot(data=df_churn, x='workclass', hue='workclass', palette='viridis', legend=False)
plt.title('Distribution of Workclass Categories')
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Observation: The plot shows that 'Private' is the most common workclass by a significant margin.
# 'Self-emp-not-inc' and 'Local-gov' are also notable, while 'Without-pay', 'Never-worked', and 'State-gov' have very low counts.
# The 'Unknown' category, which replaced '?', also shows a considerable number of individuals.

# Chart 4 : Mean Hours Per Week by Education Level

```python
plt.figure(figsize=(14, 7))
sns.barplot(x=mean_hours_per_week_by_education.index, y=mean_hours_per_week_by_education.values, hue=mean_hours_per_week_by_education.index, palette='viridis', legend=False)
plt.title('Mean Hours Per Week by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Mean Hours Per Week')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Observation: The plot indicates a general trend where higher education levels, such as 'Prof-school' and 'Doctorate', tend to have a higher mean 'hours-per-week' compared to lower education levels.
# This suggests that more advanced educational attainment may correlate with longer working hours.

##Chart 5 : Top 3 Occupations by Average Capital-Gain Plot


```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_3_occupations_capital_gain.index, y=top_3_occupations_capital_gain.values, hue=top_3_occupations_capital_gain.index, palette='viridis', legend=False)
plt.title('Top 3 Occupations by Average Capital-Gain')
plt.xlabel('Occupation')
plt.ylabel('Average Capital-Gain')
plt.tight_layout()
plt.show()
```

# Observation: The plot clearly shows that 'Prof-specialty' has the highest average capital-gain,
# followed by 'Exec-managerial' and then 'Sales'. This indicates that these professions
