7.Given stock prices, compute daily returns using vectorized NumPy operations.

```python

prices = np.array([91, 89, 115, 112, 109, 111, 81])
daily_returns = (prices[1:] - prices[:-1]) / prices[:-1]
print("Stock Prices:", prices)
print("Daily Returns:", daily_returns.round(4))
```

8. Convert an array of annual interest rates into monthly rates using vectorization.

```python
annual_rates = np.array([0.10, 0.11, 0.12, 0.13])
monthly_rates = annual_rates / 12
print("Annual Interest Rates:", annual_rates)
print("Monthly Interest Rates:", monthly_rates.round(4))
```

   9. Normalize a given array of portfolio weights so that the sum becomes 1.

```python

weights = np.array([100, 200, 300, 400,500])
normalized_weights = weights / weights.sum()
print("weights:", weights)
print("sum=1:", normalized_weights.round(4))
```

10.Create a Pandas Series from the following list: [1200, 1500, 1800, 2000, 2200]
* a.Find:
* i.mean
* ii.max
* iii.index of the max value

```python
import pandas as pd
```

```python
data_list = [1200, 1500, 1800, 2000, 2200]
series = pd.Series(data_list)
print("Pandas Series:")
display(series)
```

```python
mean_value = series.mean()
print("Mean of the Series:")
display(mean_value)
```

```python
max_value = series.max()
index_of_max = series.idxmax()
print("Maximum value of the Series:")
display(max_value)
print("Index of the maximum value:")
display(index_of_max)
```

12.Using this DataFrame:

df = pd.DataFrame({
    "Account": [101,102,103,104],
    "Balance": [5000, 15000, 7000, 12000]
})

Filter rows where:
* Balance > 8000
* Balance between 6000 and 13000

```python
print("Rows where Balance > 8000:")
display(df[df['Balance'] > 8000])
```

```python
print("Rows where Balance is between 6000 and 13000:")
display(df[(df['Balance'] >= 6000) & (df['Balance'] <= 13000)])
```
